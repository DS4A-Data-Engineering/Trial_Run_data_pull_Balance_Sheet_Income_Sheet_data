```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import plotly.express as px
import plotly.io as pio
```

```python
import fmp_api
```

```python
df = pd.read_csv('data/S&P500_bank_Tickers.csv')
```

```python
df
```

```python
def pull_request_income(ticker):
    income = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{ticker}?period=quarter&apikey={fmp_api.api_key}').json()

    return income
```

```python
def pull_request_balance(ticker):
    balance_sheet = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{ticker}?period=quarter&apikey={fmp_api.api_key}').json()

    return balance_sheet
```

```python
def pull_request_cash(ticker):
    cash_flow = requests.get(f'https://financialmodelingprep.com/api/v3/cash-flow-statement/{ticker}?period=quarter&apikey={fmp_api.api_key}').json()

    return cash_flow
```

```python
pull_request_balance('WFC')
```

```python
pull_request_cash('BLK')
```

```python
def create_report_pull(stock_ticker_list):
    report_income = []
    report_balance = []
    report_cash = []
    for t in stock_ticker_list:
        try:
            company_income = pull_request_income(t)
            company_balance = pull_request_balance(t)
            company_cash = pull_request_cash(t)
            for a, b, c in zip(company_income, company_balance, company_cash):
                report_income.append(a)
                report_balance.append(b)
                report_cash.append(c)
        except NameError:
            print("Something else went wrong")
        df_income_sheet = pd.DataFrame.from_dict(report_income)
        df_balance_sheet = pd.DataFrame.from_dict(report_balance)
        df_cash_flow = pd.DataFrame.from_dict(report_cash)
    return df_income_sheet, df_balance_sheet, df_cash_flow
```

```python
tick_list = [i for i in df['Symbol']]
tick_list
```

```python
create_report_pull(tick_list)
```

```python
x, y, z = create_report_pull(tick_list)
```
